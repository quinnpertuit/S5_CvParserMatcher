```python
#Creating education table
education_list = []

for file in collection.find():
    if('EducationHistory' in file['Resume']['StructuredXMLResume']):
        for ed in file['Resume']['StructuredXMLResume']['EducationHistory']['SchoolOrInstitution']: 
            candidate_id=level_education=career_name=start_year=end_year=degree_comments=school_name=''
            candidate_id=file['_id'] #FK
            if('Degree' in ed):
                if('@degreeType' in ed['Degree'][0]):
                    #pprint.pprint('degreeType: '+ ed['Degree'][0]['@degreeType'])
                    level_education=ed['Degree'][0]['@degreeType']

                if('DegreeMajor' in ed['Degree'][0]):
                    if('Name' in ed['Degree'][0]['DegreeMajor'][0]):
                        #pprint.pprint('degreeMajor: '+ed['Degree'][0]['DegreeMajor'][0]['Name'][0])
                        career_name=ed['Degree'][0]['DegreeMajor'][0]['Name'][0]

                if('DatesOfAttendance' in ed['Degree'][0]):
                    if('StartDate' in ed['Degree'][0]['DatesOfAttendance'][0]):
                        #pprint.pprint(ed['Degree'][0]['DatesOfAttendance'][0]['StartDate'])
                        start_year=ed['Degree'][0]['DatesOfAttendance'][0]['StartDate']

                    if('EndDate' in ed['Degree'][0]['DatesOfAttendance'][0]):
                        #pprint.pprint(ed['Degree'][0]['DatesOfAttendance'][0]['EndDate'])
                        end_year=ed['Degree'][0]['DatesOfAttendance'][0]['EndDate']

                if('Comments' in ed['Degree'][0]):
                    #pprint.pprint(ed['Degree'][0]['Comments'])
                    degree_comments=ed['Degree'][0]['Comments']

            if('School' in ed):
                if('SchoolName' in ed['School'][0]):
                    #pprint.pprint('SchoolName: '+ed['School'][0]['SchoolName'])
                    school_name=ed['School'][0]['SchoolName']

            #print("-------------")
            education_list.append([candidate_id,level_education,career_name,school_name,start_year,end_year,degree_comments])
```

```python
df = pd.DataFrame(education_list, columns=['candidate_id','level_education','career_name','school_name','start_year','end_year','degree_comments'])
df
```

```python
#Creating experience table

#Creating experience table
#CandidateID
#EmployerOrgName
#@positionType
#Title
#Description
#StartDate
#EndDate



experience_list = []

for file in collection.find():
    if('EmploymentHistory' in file['Resume']['StructuredXMLResume']):
        for ex in file['Resume']['StructuredXMLResume']['EmploymentHistory']['EmployerOrg']: 
            candidate_id=EmployerOrgName=positionType=Title=Description=StartDate=EndDate=''
            candidate_id=file['_id'] #FK
            if('PositionHistory' in ex):
                if('OrgName' in ex['PositionHistory'][0]):
                    EmployerOrgName=ex['PositionHistory'][0]['OrgName']['OrganizationName']
                    
                if('@positionType' in ex['PositionHistory'][0]):
                    positionType=ex['PositionHistory'][0]['@positionType']
                    
                if('Title' in ex['PositionHistory'][0]):
                    Title=ex['PositionHistory'][0]['Title']
                    
                if('Description' in ex['PositionHistory'][0]):
                    Description=ex['PositionHistory'][0]['Description']
                    
                if('StartDate' in ex['PositionHistory'][0]):
                    StartDate=ex['PositionHistory'][0]['StartDate']
                    
                if('EndDate' in ex['PositionHistory'][0]):
                    EndDate=ex['PositionHistory'][0]['EndDate']            
            experience_list.append([candidate_id,EmployerOrgName,positionType,Title,Description,StartDate,EndDate])
```

```python
from pymongo import MongoClient
from random import randint
#Step 1: Connect to MongoDB - Note: Change connection string as needed
client = MongoClient(port=27017)
db=client.business
#Step 2: Create sample data
names = ['Kitchen','Animal','State', 'Tastey', 'Big','City','Fish', 'Pizza','Goat', 'Salty','Sandwich','Lazy', 'Fun']
company_type = ['LLC','Inc','Company','Corporation']
company_cuisine = ['Pizza', 'Bar Food', 'Fast Food', 'Italian', 'Mexican', 'American', 'Sushi Bar', 'Vegetarian']
for x in range(1, 501):
    business = {
        'name' : names[randint(0, (len(names)-1))] + ' ' + names[randint(0, (len(names)-1))]  + ' ' + company_type[randint(0, (len(company_type)-1))],
        'rating' : randint(1, 5),
        'cuisine' : company_cuisine[randint(0, (len(company_cuisine)-1))] 
    }
    #Step 3: Insert business object directly into MongoDB via isnert_one
    result=db.reviews.insert_one(business)
    #Step 4: Print to the console the ObjectID of the new document
    print('Created {0} of 500 as {1}'.format(x,result.inserted_id))
#Step 5: Tell us that you are done
print('finished creating 500 business reviews')
```
